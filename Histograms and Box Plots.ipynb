# Histogram

Sintaxis

<b>hist(x)</b><br>
'x' : Data values of a single variable.

<li>The below example simulates 1000 percentage values from a normal distribution with mean 60 and standard deviation 10.</li>

<li>Then the histogram of percentage values is plotted.</li>


```python
from matplotlib import pyplot as plt
import numpy as np
np.random.seed(100)
x = 60 + 10*np.random.randn(1000)
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
ax.set(title="Distribution of Student's Percentage",
      ylabel='Count', xlabel='Percentage')
ax.hist(x)
plt.show()
```

Common Parameters of 'hist'

<b>color:</b> Sets the color of bars.

<b>bins:</b> Sets the number of bins to be used.

<b>normed:</b> Sets to True where bins display fraction and not the count.

You can also create more bins and show bin count as a fraction (as specified in the following example).

```python
import numpy as np
np.random.seed(100)
x = 60 + 10*np.random.randn(1000)
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
ax.set(title="Distribution of Student's Percentage",
      ylabel='Proportion', xlabel='Percentage')
ax.hist(x, color='blue', bins=30, density=True)
plt.show()
```

# Box Plots

<b>Box plots are also used to visualize the spread of data.</b><br>
<b>Box plots are used to compare distributions.</b><br>
<b>Box plots can also be used to detect outliers.</b><br>

### Sintaxis

<b>boxplot(x)</b> <br>
'x' : list of values or list of list of values.

<li>The above image displays box plot of Percentages obtained by 1000 Students of a class.</li>

```python
import numpy as np
np.random.seed(100)
x = 50 + 10*np.random.randn(1000)
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
ax.set(title="Box plot of Student's Percentage",
      xlabel='Class', ylabel='Percentage')
ax.boxplot(x)
plt.show()
```

### Common Parameters of 'boxplot'

<b>labels:</b> Sets the labels for box plots.

<b>notch:</b> Sets to True if notches need to be created around the median.

<b>bootstrap:</b> Number set to indicate that notches around the median are bootstrapped.

<b>vert:</b> Sets to False for plotting Box plots horizontally.

<li>Box plot of Student Percentages can be redrawn by setting notch, bootstrap and labels using the below-shown expression.</li>
<b>ax.boxplot(x, labels=['A'], notch=True, bootstrap=10000)</b>

```python
import numpy as np
np.random.seed(100)
x = 50 + 10*np.random.randn(1000)
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
ax.set(title="Box plot of Student's Percentage",
      xlabel='Class', ylabel='Percentage')
ax.boxplot(x, labels=['A'], notch=True, bootstrap=10000)
plt.show()
```

## Plotting Multiple Boxplots

List of data values can be passed as an argument for plotting multiple box plots as shown in below code snippet.

```python
import numpy as np
np.random.seed(100)
x = 50 + 10*np.random.randn(1000)
y = 70 + 25*np.random.randn(1000)
z = 30 + 5*np.random.randn(1000)
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
ax.set(title="Box plot of Student's Percentage",
      xlabel='Class', ylabel='Percentage')
ax.boxplot([x, y, z], labels=['A', 'B', 'C'], notch=True, bootstrap=10000)
plt.show()
```

### Plotting Boxplots Horizontally

Box plots are plotted horizontally by setting vert to False, as shown in the below code snippet.

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)

ax.set(title="Box plot of Student's Percentage",
      xlabel='Percentage', ylabel='Class')
ax.boxplot([x, y, z], labels=['A', 'B', 'C'], vert=False, notch=True, bootstrap=10000)
plt.show()
```
