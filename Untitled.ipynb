# Matplotlib Styles

<li>matplotlib.pyplot comes with a lot of styles. Based on the chosen style, the display of figure changes.</li>
<li>You can view various styles available in pyplot by running the following commands.</li>

```python
import matplotlib.pyplot as plt
print(plt.style.available)
```


<li>A specific style can be invoked with either of the two expressions shown below.</li>

<b>plt.style.use('ggplot')</b>

or

<b>plt.style.context('ggplot')</b>

Using the later expression with a keyword, with is recommended.



```python
with plt.style.context('ggplot'):
    fig = plt.figure(figsize=(8,6))
    ax = fig.add_subplot(111)
    ax.set(title='Avg. Daily Temperature of Jan 2018',
      xlabel='Day', ylabel='Temperature (in deg)',
      xlim=(0, 30), ylim=(25, 35))
    days = [1, 5, 8, 12, 15, 19, 22, 26, 29]
    temp = [29.3, 30.1, 30.4, 31.5, 32.3, 32.6, 31.8, 32.4, 32.7]
    ax.plot(days, temp, color='green', linestyle='--', linewidth=3)
    plt.show()
```

## Creating a Custom Style

<ul>A style sheet is a text file having extension .mplstyle.</ul>

<ul>All custom style sheets are placed in a folder, stylelib, present in the config directory of matplotlib.</ul>

<ul>Use the below expression for knowing the Config folder.</ul>




```python
import matplotlib
print(matplotlib.get_configdir())
```

Now, create a file mystyle.mplstyle with the below-shown contents and save it in the folder <matplotlib_configdir/stylelib/.
<br>
<b>
axes.titlesize : 24<br>
axes.labelsize : 20<br>
lines.linewidth : 8<br>
lines.markersize : 10<br>
xtick.labelsize : 16<br>
ytick.labelsize : 16<br>
</b>

Reload the matplotlib library with the subsequent expression.

<b>matplotlib.style.reload_library()</b>
