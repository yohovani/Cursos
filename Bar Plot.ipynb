```python
def barplot_of_iris_sepal_length():

    # Write your functionality below
    fig = plt.figure(figsize=(9,7))
    ax = fig.add_subplot(111)
    ax.set(title='Mean Sepal Length (cm)',
      xlabel='Species', ylabel='Sepal Length (cm)', xlim = (0,3), ylim = (0,9))
    species = ['setosa','versicolor','virginica']
    index =[0.4,1.4,2.4]
    sepal_len = [6.01,6.94,7.69]
    ax.set_xticks(index)
    ax.set_xticklabels(species)
    ax.bar(index, sepal_len, color='red', width=0.4, edgecolor='black')
    ax.legend()
    plt.show()
```

## Excersice 2

```python
def barplot_of_iris_measurements():

    # Write your functionality below
    fig = plt.figure(figsize=(9,7))
    ax = fig.add_subplot(111)
    ax.set(title='Mean Measurements of Iris Species',
      xlabel='Species', ylabel='Iris Measurements (cm)', xlim = (0.5,3.5), ylim = (0,10))

    sepal_len = [6.01,6.94,7.69]
    sepal_wd = [4.42,3.77,3.97]
    petal_len = [2.46,5.26,6.55]
    petal_wd = [1.24,2.33,3.03]
    species = ['setosa','versicolor','virginica']
    species_index1 =[0.7,1.7,2.7]
    species_index2 = [0.9,1.9,2.9]
    species_index3 = [1.1,2.1,3.1]
    species_index4 = [1.3,2.3,3.3]
    ax.set_xticks([1.0,2.0,3.0])
    ax.set_xticklabels(species)
    ax.bar(species_index1, sepal_len, color='m', edgecolor='gray', width=0.2, label="Sepal Length")
    ax.bar(species_index2, sepal_wd, color='y', edgecolor='grey', width=0.2, label="Sepal Width")
    ax.bar(species_index3, petal_len, color='c', edgecolor='grey', width=0.2, label="Petal Length")
    ax.bar(species_index4, petal_wd, color='orange', edgecolor='grey', width=0.2, label="Petal Width")
    ax.legend()
    plt.show()
```

```python
barplot_of_iris_measurements()
```

```python
def hbarplot_of_iris_petal_length():

    # Write your functionality below
    fig = plt.figure(figsize=(15,5))
    ax = fig.add_subplot(111)
    ax.set(title='Mean Measurements of Iris Species',
      ylabel='Species', label='Petal Length (cm)')
    species = ['setosa','versicolor','virginica']
    petal_len = [2.57,5.49,6.37]
    index = [0.1,1.1,2.1]
    ax.set_yticks([0.10,1.10,2.10])
    ax.set_yticklabels(species)
    ax.barh(index, petal_len, height=0.4, color='m', edgecolor='c')
    ax.legend()
    plt.show()
```

```python
hbarplot_of_iris_petal_length()
```
