# Matplotlib Styles

<li>matplotlib.pyplot comes with a lot of styles. Based on the chosen style, the display of figure changes.</li>
<li>You can view various styles available in pyplot by running the following commands.</li>

```python
import matplotlib.pyplot as plt
print(plt.style.available)
```


<li>A specific style can be invoked with either of the two expressions shown below.</li>

<b>plt.style.use('ggplot')</b>

or

<b>plt.style.context('ggplot')</b>

Using the later expression with a keyword, with is recommended.



```python
with plt.style.context('ggplot'):
    fig = plt.figure(figsize=(8,6))
    ax = fig.add_subplot(111)
    ax.set(title='Avg. Daily Temperature of Jan 2018',
      xlabel='Day', ylabel='Temperature (in deg)',
      xlim=(0, 30), ylim=(25, 35))
    days = [1, 5, 8, 12, 15, 19, 22, 26, 29]
    temp = [29.3, 30.1, 30.4, 31.5, 32.3, 32.6, 31.8, 32.4, 32.7]
    ax.plot(days, temp, color='green', linestyle='--', linewidth=3)
    plt.show()
```

```python
import matplotlib
print(matplotlib.get_configdir())
```

## matplotlibrc file

<li>matplotlib uses all the settings specified in matplotlibrc file.</li>

<li>These settings are known as rc settings or rc parameters.</li>

<li>For customization, rc settings can be altered in the file or interactively.</li>

<li>The location of active matplotlibrc file used by matplotlib can be found with below expression.</li>


```python
import matplotlib
matplotlib.matplotlib_fname()
```



    All rc settings, present in matplotlibrc file are stored in a dictionary named matplotlib.rcParams.

    Any settings can be changed by editing values of this dictionary.

    For example, if you want to change linewidth and color, the following expressions can be used.

<b>
import matplotlib as mpl<br>
mpl.rcParams['lines.linewidth'] = 2<br>
mpl.rcParams['lines.color'] = 'r'<br></b>

## Hands-Ons

```python
import matplotlib.pyplot as plt
def generate_plot_with_style1():

    # Write your functionality below
    plt.style.use('ggplot')
    fig = plt.figure(figsize=(9,7))
    ax = fig.add_subplot(111)
    ax.set(title='Measurements of Iris Species',
           xlabel='Species', ylabel='Iris Mesuraments(cm)',
           xlim=(0.5, 3.7), ylim=(0, 10))
    sepal_len = [6.01,6.94,7.59]
    sepal_wd = [4.42,3.77,3.97]
    petal_len = [2.46, 5.26, 6.55]
    petal_wd = [1.24,2.33,3.03]
    species = ['setosa','versicolor','virginica']
    species_index1 = [0.8,1.8,2.8]
    species_index2= [1.0,2.0,3.0]
    species_index3= [1.2,2.2,3.2]
    species_index4= [1.4,2.4,3.4]
    ax.set_xticks([1.0,2.0,3.0])
    ax.set_xticklabels(species)
    ax.bar(species_index1, sepal_len, width=0.2)
    ax.bar(species_index2, sepal_wd, width=0.2)
    ax.bar(species_index3, petal_len, width=0.2)
    ax.bar(species_index4, petal_wd, width=0.2)
    ax.legend()
    return fig
```

```python
generate_plot_with_style1()
```
